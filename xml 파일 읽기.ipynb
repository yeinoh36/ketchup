```python
!pip install xmltodict

import requests
import pprint
import json
import xmltodict

url = 'http://www.djtc.kr/OpenAPI/service/CCTVSVC/getCCTVData01?serviceKey=fGEu6dduFABfox8Ku3PcWiiXT7pGoSmwGU3MePt4JWI/ZQyOkcGKLMFZjRnrJRwlZrXv8go4Fsa76877nBgXjQ==&pageNo=1&numOfRows=100&type=json'

response = requests.get(url)
contents = response.text


# XML을 OrderedDict로 파싱
xml_dict = xmltodict.parse(contents)

# "items" 아래의 "item" 추출
body = xml_dict['response']['body']['items']['item']

# 출력
pprint.pprint(body)

# pandas import
import pandas as pd

# DataFrame으로 만들기
dataframe = pd.DataFrame(body)

# 출력
dataframe

dataframe.to_csv('대전교통공사_CCTV정보', index=False)
```
