```python
from geopy.distance import geodesic
# Get user input for the search radius in meters
search_radius = float(input('반경 몇 미터 내의 CCTV를 검색할까요?: '))

# 인근 지정된 반경 내의 CCTV 찾기
nearby_cctv = pd.DataFrame(columns=merged_cctv_data.columns)

for index, cctv_row in merged_cctv_data.iterrows():
    # Check if both latitude and longitude are not NaN
    if not pd.isnull(cctv_row['crdntY']) and not pd.isnull(cctv_row['crdntX']):
        cctv_latitude = cctv_row['crdntY']
        cctv_longitude = cctv_row['crdntX']
    else:
        continue

    cctv_location = (cctv_latitude, cctv_longitude)
    distance = geodesic((user_latitude, user_longitude), cctv_location).meters

    if distance <= search_radius:
        nearby_cctv = pd.concat([nearby_cctv, pd.DataFrame(cctv_row).transpose()], ignore_index=True)

    # NaN 값을 가진 열 제거
nearby_cctv = nearby_cctv.dropna(axis=1, how='all')

print(f'인근 {search_radius}m 반경 내의 CCTV 목록:')
print(nearby_cctv)


```
